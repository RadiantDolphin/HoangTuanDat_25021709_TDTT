```python
#W5A1
def Max_Of_Two(a,b):
    return max(a,b)

a,b = map(int,input("nhap a b").split())
Max_Of_Two(a,b)
```

```python
#W5A2
def swap(a,b):
    return tuple([b,a])
a,b = map(int,input("nhap a b").split())
swap(a,b)
```

```python
#W5A3
def check_prime(x):
    if x < 2:
        return False
    for i in range(2,round(x/2)+1):
        if x % i == 0:
            return False
    return True
x = int(input("nhap x:"))
check_prime(x)
```

```python
#W5A4
def is_perfect(x):
    total = 0
    for i in range(1,round(x/2)+1):
        if x % i == 0:
            total += i
    return x == total
    
x = int(input("nhap x:"))
is_perfect(x)
```

```python
#W5A5
def search(user_input,k):
    number_list = list(map(int, user_input))

    if k in number_list:
        return number_list.index(k)
    else:
        return -1


user_input = input("nhap a b").split()
k = int(input("nhap k"))
search(user_input,k)
```

```python
#W5A7
def calc(x,y,op):
    if op == "+":
        return x + y
    elif op == "-":
        return x - y
    elif op == x * y:
        return x * y
    elif op == "/":
        return x / y
    else:
        return ":("
x,y = map(float,input("nhap x y").split())
op = input("nhap operator")
calc(x,y,op)
```

```python
#W5A8
def hamming(x,y):
    n = x ^ y
    return n.bit_count()
x, y = map(int,input().split())
hamming(x,y)
```
