```python
#B4
def triangle(h):
    for row in range(1,h+1):
        if row == 1:
            print(f'{(h-row)*" "}*')
        elif row == h:
            print(f"*{'*'*(2*row-3)}*")
        else:
            print(f'{(h-row)*" "}*{' '*(2*row-3)}*')
try:
    h = int(input("nhap so nguyen duong h: "))
    if h > 0: triangle(h)
    else: print("non-positive integer error")
except ValueError:
    print("dfkhfeuroidfhjk")
```

```python
#B8
#W4A14

def factors(n):
    factors = []
    for i in range (1,n+1):
        if n % i == 0:
            factors.append(i)
    return factors
try:
    user_input = input("nhap a, b:")
    a,b = map(int,user_input.split())
    if a <= 0 or b <= 0:
        print("value error")
    else:
        f_of_a = factors(a)
        f_of_b = factors(b)
        for factor in f_of_a:
            if factor in f_of_b:
                print(factor,end = " ")
except ValueError:
    print('value error')
```

```python
#B12
import math
try:
    n = int(input("nhap so nguyen duong n: "))
    if n > 0:
        k = 1
        for _ in range(0,int(math.sqrt(n))+1):
            if k*(k+1)/2 >= n:
                k -= 1
                break
            else:
                k += 1
        print("k =",k)
    else: print("non-positive integer error")
except ValueError:
    print("not an integer error")
```
